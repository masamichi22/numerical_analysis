```python
import numpy as np
import matplotlib.pyplot as plt
```

これは区間積分
\begin{equation}
\int_{0}^{\pi}\mathrm{sin}x dx = [\mathrm{cos} x]_{0}^{\pi} = 2
\end{equation}
を数値計算で求めるPythonプログラムです

```python
x = np.linspace(0, 2*np.pi, 10000)
plt.plot(x, np.sin(x), label='sin($x$)')
plt.xlabel('$x$')
plt.ylabel('sin($x$)')
plt.grid()
plt.show()
```

```python
def func(x):
    return np.sin(x)
def daikei_koshiki(func, init, end, divide_num):
    h = (end - init) / divide_num # 台形の高さに相当するもの
    x = np.linspace(init, end, divide_num+1)
    result = 0
    for i in range(1, len(x)):
        result += (func(x[i-1]) + func(x[i])) * h / 2
        
    return result

print(daikei_koshiki(func, 0, np.pi, 10000))
```

解析的に解いた値と，数値計算の値がほぼ一致しました
