## 力学系のモデリング

### BMS System

Problem definition is below.
\begin{eqnarray}
m \frac{d^2x(t)}{dt^2} = -c\frac{dx(t)}{dt} - kx(t)\\
\therefore \frac{d^2x(t)}{dt^2} + \frac{c}{m}\frac{dx(t)}{dt} + \frac{k}{m}x(t) = 0
\end{eqnarray}
(m: mass, c: damper coefficient, k: spring coefficient)

```python
# バネマスダンパ系を数値積分（前進オイラー法）で解く
import numpy as np
import matplotlib.pyplot as plt
# 判別式より 
m = 1.0
c = 1.0
k = 0.2
timestep = 0.001
total_time = 10
x_init = 1.0
v_init = 0.0

time_steps = int(total_time / timestep)
t = np.linspace(0, total_time, time_steps)

x = np.zeros(time_steps)
v = np.zeros(time_steps)
x[0] = x_init
v[0] = v_init
# オイラー法による積分
for i in range(1, time_steps):
    x[i] = x[i-1] + v[i-1] * timestep
    v[i] = v[i-1] - (k/m) * x[i-1] * timestep - (c/m) * v[i-1] * timestep

plt.figure(figsize=(12, 4))
plt.plot(t, x)
plt.xlabel('time[sec]')
plt.ylabel('amplitude[m]')
plt.grid()
plt.show()
```
